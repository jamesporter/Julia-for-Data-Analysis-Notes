# Adding data we generate

```julia
function sim_step(current)
    dx, dy = rand(((1,0), (-1,0), (0,1), (0,-1)))
    return (x=current.x + dx, y=current.y + dy)
end
```

```julia
?rand
```

```julia
for i in 1:10 
    println(rand((1,2,3)))
end
```

```julia
walk = DataFrame(x=0, y=0)
```

```julia
for _ in 1:100000
    current = walk[end,:]
    push!(walk, sim_step(current))
end
```

```julia
using Plots
```

```julia
plot(walk.x, walk.y;
            legend=false)
```

```julia
using Serialization
```

```julia
 serialize("walk.bin", walk)
```

```julia
w = deserialize("walk.bin")
```
